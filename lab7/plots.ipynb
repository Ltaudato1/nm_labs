# Решение краевой задачи методом конечных-разностей

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Зависимость погрешности решения от шага интегрирования

```python
data = pd.read_csv('data/part_error.csv')

plt.figure(figsize=(12,6))
plt.xlabel('Шаг интегрирования')
plt.ylabel('Погрешность')
plt.xscale('log', base=2)
plt.yscale('log', base=2)

error = data['error']
partitions = data['part']
plt.plot(partitions, error, color='g')

plt.grid(True)
plt.show()
```
