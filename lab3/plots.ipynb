```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Графики функций

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
xticks = np.arange(-2 * np.pi, 2 * np.pi + 1, np.pi)
xticks_labels = [r"$-2\pi$", r"$-\pi$", r"$0$", r"$\pi$", r"$2\pi$"]
```

## $y = 2sin(x)$

```python
def smooth(x):
    return 2 * np.sin(x)
```

```python
y_smooth = smooth(x)

plt.plot(x, y_smooth, color='r')

plt.xlabel('x')
plt.ylabel('y')

ax = plt.gca()
ax.set_xticks(xticks)
ax.set_xticklabels(xticks_labels)

plt.grid(True)
plt.show()
```

## $y = |x - e|cos(x)$

```python
def breakdown(x):
    return abs(x - np.e) * np.cos(x)
```

```python
y_breakdown = breakdown(x)
xticks = np.arange(-2 * np.pi, 2 * np.pi + 1, np.pi)

plt.plot(x, y_breakdown, color='r')

plt.xlabel('x')
plt.ylabel('y')

ax = plt.gca()
ax.set_xticks(xticks)
ax.set_xticklabels(xticks_labels)

plt.grid(True)
plt.show()
```

# Зависимость погрешности от числа разбиений

```python
plt.xlabel('Число разбиений')
plt.ylabel('Погрешность')
plt.xscale('log', base=10)
plt.yscale('log', base=10)

data = pd.read_csv("data/part_error_smooth.csv")
error = data['error']
partitions = data['part']
plt.plot(partitions, error, color='g', label='Гладкая функция')

data = pd.read_csv("data/part_error_breakdown.csv")
error = data['error']
partitions = data['part']
plt.plot(partitions, error, color='r', label='Функция с разрывом производной')

plt.grid(True)
plt.legend()
plt.show()
```
