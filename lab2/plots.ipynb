```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Графики интерполяционных сплайнов

## $y = 3x - 14 + e^{-x}$

```python
def smooth(x):
    return 3 * x - 14 + np.exp(-x)
```

```python
x = np.linspace(-4, 5, 1000)
y_smooth = smooth(x)

plt.plot(x, y_smooth, color='r', label='Гладкая функция')

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

### Чебышевская сетка

```python
data = pd.read_csv("data/chebyshev_smooth_2.csv")
x = data['x']
y_function = smooth(x)

plt.plot(x, y_function, color='black', linestyle='--', label='$f(x)$', linewidth=3)
colors = ['r', 'g', 'b', 'orange']

for i in range(2, 9, 2):
    name = "data/chebyshev_smooth_" + str(i) + ".csv"
    label = str(i)
    if (i < 5):
        label += ' узла'
    else:
        label += ' узлов'
    data = pd.read_csv(name)
    y_inter = data['y']
    plt.plot(x, y_inter, color=colors[i // 2 - 1], label=label)

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

### Равномерная сетка

```python
data = pd.read_csv("data/uniform_smooth_2.csv")
x = data['x']
y_function = smooth(x)

plt.plot(x, y_function, color='black', linestyle='--', label='$f(x)$', linewidth=3)
colors = ['r', 'g', 'b', 'orange']

for i in range(2, 9, 2):
    name = "data/uniform_smooth_" + str(i) + ".csv"
    label = str(i)
    if (i < 5):
        label += ' узла'
    else:
        label += ' узлов'
    data = pd.read_csv(name)
    y_inter = data['y']
    plt.plot(x, y_inter, color=colors[i // 2 - 1], label=label)

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

## $y = |x + ln(2)|sin(x) + 1$

```python
def breakdown(x):
    return abs(x + np.log(2)) * np.sin(x) + 1
```

### Чебышевская сетка

```python
data = pd.read_csv("data/chebyshev_breakdown_2.csv")
x = data['x']
y_function = breakdown(x)

plt.plot(x, y_function, color='black', linestyle='--', label='$g(x)$', linewidth=3)
colors = ['r', 'g', 'b', 'orange']

for i in range(2, 9, 2):
    name = "data/chebyshev_breakdown_" + str(i) + ".csv"
    label = str(i)
    if (i < 5):
        label += ' узла'
    else:
        label += ' узлов'
    data = pd.read_csv(name)
    y_inter = data['y']
    plt.plot(x, y_inter, color=colors[i // 2 - 1], label=label)

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

### Равномерная сетка

```python
data = pd.read_csv("data/uniform_breakdown_2.csv")
x = data['x']
y_function = breakdown(x)

plt.plot(x, y_function, color='black', linestyle='--', label='$g(x)$', linewidth=3)
colors = ['r', 'g', 'b', 'orange']

for i in range(2, 9, 2):
    name = "data/uniform_breakdown_" + str(i) + ".csv"
    label = str(i)
    if (i < 5):
        label += ' узла'
    else:
        label += ' узлов'
    data = pd.read_csv(name)
    y_inter = data['y']
    plt.plot(x, y_inter, color=colors[i // 2 - 1], label=label)

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

# Зависимость погрешности от количества узлов

## $y = 3x - 14 + e^{-x}$

```python
data = pd.read_csv("data/nodes_error_chebyshev_smooth.csv")
nodes = data['nodes']
error = data['error']


plt.plot(nodes, error, color='r', label='Чебышевская сетка')

data = pd.read_csv("data/nodes_error_uniform_smooth.csv")
nodes = data['nodes']
error = data['error']


plt.plot(nodes, error, color='b', label='Равномерная сетка')
plt.xlabel('nodes')
plt.ylabel('error')
plt.xticks(range(2, 103, 10))
plt.grid(True)
plt.yscale('log')
plt.legend()
plt.show()
```
