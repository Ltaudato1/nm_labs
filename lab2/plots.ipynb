```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Графики интерполяционных сплайнов

## $y = 3x - 14 + e^{-x}$

```python
def smooth(x):
    return 3 * x - 14 + np.exp(-x)
```

### Чебышевская сетка

```python
data = pd.read_csv("data/chebyshev_smooth_plot.csv")
x = data['x']
y_spline = data['y']
y_function = smooth(x)

plt.plot(x, y_spline, color='r')
plt.plot(x, y_function, color='black', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

### Равномерная сетка

```python
data = pd.read_csv("data/uniform_smooth_plot.csv")
x = data['x']
y_spline = data['y']
y_function = smooth(x)

plt.plot(x, y_spline, color='r')
plt.plot(x, y_function, color='black', linestyle='--')

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

## $y = 3x - 14 + e^{|x|}$

```python
def breakdown(x):
    return 3 * x - 14 + np.exp(abs(x))
```

### Чебышевская сетка

```python
data = pd.read_csv("data/chebyshev_breakdown_plot.csv")
x = data['x']
y_spline = data['y']
y_function = breakdown(x)

plt.plot(x, y_spline, color='r')
plt.plot(x, y_function, color='black', linestyle='--')
plt.ylim(-20, 30)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

### Равномерная сетка

```python
data = pd.read_csv("data/uniform_breakdown_plot.csv")
x = data['x']
y_spline = data['y']
y_function = breakdown(x)

plt.plot(x, y_spline, color='r')
plt.plot(x, y_function, color='black', linestyle='--')
plt.ylim(-20, 30)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

# Зависимость погрешности от количества узлов

## $y = 3x - 14 + e^{-x}$

```python
data = pd.read_csv("data/nodes_error_chebyshev_smooth.csv")
nodes = data['nodes']
error = data['error']


plt.plot(nodes, np.log10(error), color='r', label='Чебышевская сетка')

data = pd.read_csv("data/nodes_error_uniform_smooth.csv")
nodes = data['nodes']
error = data['error']


plt.plot(nodes, np.log10(error), color='b', label='Равномерная сетка')
plt.xlabel('nodes')
plt.ylabel('lg(error)')
plt.xticks(range(2, 103, 10))
plt.grid(True)
plt.legend()
plt.show()
```
