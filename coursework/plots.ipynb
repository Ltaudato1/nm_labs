## Зависимость погрешности решения ОДУ от шага интегрирования

```python
data = pd.read_csv('data/part_error.csv')

plt.figure(figsize=(10,6))
plt.xlabel('Шаг интегрирования')
plt.ylabel('Погрешность')
plt.xscale('log', base=2)
plt.yscale('log', base=2)

error_adams = data['error_adams']
error_rk = data['error_rk']
partitions = data['part']
plt.plot(partitions, error_adams, 'r-o', label='Метод Адамса')
plt.plot(partitions, error_rk, 'b-o', label='Метод Рунге-Кутты')

plt.grid(True)
plt.legend()
plt.show()
```
