## Зависимость погрешности решения ОДУ от шага интегрирования

```python
data = pd.read_csv('data/part_error.csv')
partitions = data['part']
error_rk = data['error_rk']
error_adams = data['error_adams']

k_min, k_max = -40, 10
y_ticks = [2**k for k in range(k_min, k_max+1, 5)]

fig, ax = plt.subplots(figsize=(10,6))
ax.set_xscale('log', base=2)
ax.set_yscale('log', base=2)

ax.set_yticks(y_ticks)

ax.set_yticklabels([f'$2^{{{k}}}$' for k in range(k_min, k_max+1, 5)])

ax.plot(partitions, error_adams, 'r-o', label='Метод Адамса')
ax.plot(partitions, error_rk,    'b-o', label='Метод Рунге–Кутты')

ax.grid(which='both', linestyle='--', linewidth=0.5)
ax.legend()
plt.show()
```
